### Data Set

- In case you don't know what pokémon are, they're creatures from the pokémon series of video games. They have 6 stats and 1-2 elemental types they are associated with. I want to see 
- Our dataset is a pruned/preprocessed list of just about every pokémon as of November 2023
- Each sample pokémon includes:
    1) Name (will not be used for any calculation)
    2) Primary Elemental Type
    3) Secondary Elemental Type (if applicable)
    4) HP Stat
    5) Attack Stat (for attacks that make physical contact)
    6) Defense Stat (for defense against physical attacks)
    7) Special Attack Stat (for attacks that don't make contact ex: attacking with a lightning bolt)
    8) Special Defense Stat (for defense against non-physical attacks)
    9) Speed Stat
    10) Stat Total
- Our data source comes from two places:
    1) https://www.kaggle.com/datasets/hamzacyberpatcher/data-of-1010-pokemons/
    2) https://pokemondb.net

### Preprocessing 

- Now I wanted to include every pokémon to make sure my data set was large enough
- My initial set from kaggle included each pokémon once, however some poké have alternate forms with different stats or elemental typings, so I had to add those
- I also removed most data from the initial kaggle data set that doesn't relate to the stats or typing
- All pokémon's alternate forms are included except for forms that are only temporary (in battle transformations and such)
- These were all manually removed or added myself using data from https://pokemondb.net

### Will Nearest Neighbors show different patterns?
- Perhaps a linear relationship is less accurate?
- Let's see if maybe there are large jumps in stats (maybe similar to archetypes or thresholds)

```python
def neighbors(df, s, n):
    '''Finds the n samples in df nearest to the sample s.
       
       Arguments:
       arr - an Nx(M+1) dataframe containing N samples, each containing M features and 1 response
       s  - a sample: a 1xM numpy array of features
       n  - number of nearest neighbors to find
       
       Returns:
            an nx(M+1) dataframe containing the n nearest samples
    '''
    
    # Grab the features in a numpy array
    N = df.shape[0]
    M = df.shape[1]
    arr = df.values[0:N, 1:M+1]
    
    # Store Euclidean distance from s to each neighbor in an array
    # (Use the calculation in the scratch cell)
    dist = np.sqrt( np.sum(((s-arr)**2), axis=1) )
    
    # Add the distances as a new column called 'distance' to the dataframe
    df['distance'] = dist
    
    # Sort the samples by distance to s.
    # Drop the distance column from the result, then retain first n samples (by default, drop returns a copy)
    retval = df.sort_values(by=['distance']).drop(columns=['distance'])
    
    # remove new column from df (with inplace=True, this modifies df and returns None)
    df.drop(columns=['distance'], inplace=True)
    
    return retval.head(n)
```
